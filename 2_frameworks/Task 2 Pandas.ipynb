#### 21. Замените все пропущенные значения, на среднее по столбцу(группе)
```python
df = pd.DataFrame({'group': list('aabbabbbabab'),
                       'value': [1, 2, 3, np.nan, 2, 3, 
                                 np.nan, 1, 7, 3, np.nan, 8]})
```

````python
df = pd.DataFrame({'group': list('aabbabbbabab'),
                       'value': [1, 2, 3, np.nan, 2, 3, 
                                 np.nan, 1, 7, 3, np.nan, 8]})
df_nona = df.dropna(axis = 0, how='any')
means = df_nona.groupby('group').mean().loc[:, 'value']
df.loc[df.value.isna(), 'value'] = np.where(df.loc[df.value.isna(), 'group'] == 'a', means[0], means[1])
df
````

````python
df = pd.DataFrame({'From_To': ['LoNDon_paris', 'MAdrid_miLAN', 'londON_StockhOlm', 
                               'Budapest_PaRis', 'Brussels_londOn'],
              'FlightNumber': [10045, np.nan, 10065, np.nan, 10085],
              'RecentDelays': [[23, 47], [], [24, 43, 87], [13], [67, 32]],
                   'Airline': ['KLM(!)', '<Air France> (12)', '(British Airways. )', 
                               '12. Air France', '"Swiss Air"']})
````

````python
 df
````

````python
def get_from(s):
    fr = s.split('_')[0].title()
    return fr

def get_to(s):
    to = s.split('_')[1].title()
    return to

to = df.From_To.apply(get_to)
fr = df.From_To.apply(get_from)

df['From'] = fr
df['To'] = to
df
````

````python
df.drop(['From', 'To'], axis = 1, inplace=True)
df
````

````python
df['From'] = [s.split('_')[0].capitalize() for s in df['From_To']]
df['To'] = [s.split('_')[1].capitalize() for s in df['From_To']]
df
````
